# 1.Creation de la data base 

```python

cursor.execute("USE SADOP_BDD")
print("Database selected")
```

## 1.2Users Tabel

```python
cursor.execute("""
CREATE TABLE user (
    user_id INT PRIMARY KEY,
    full_name VARCHAR(100),
    email VARCHAR(120),
    country CHAR(2),
    signup_date DATE
)
""")

print("âœ… users table created")
```

## 1.3 Accounts table

```python
cursor.execute("""
CREATE TABLE accounts (
    account_id INT PRIMARY KEY,
    user_id INT,
    account_type ENUM('SAVINGS','CHECKING','BUSINESS'),
    balance DECIMAL(12,2),
    created_at DATE,
    FOREIGN KEY (user_id) REFERENCES user(user_id)
)
""")

print("âœ… accounts table created")
```

## 1.4 Transactions table 

```python
cursor.execute("""
CREATE TABLE transactions (
    transaction_id INT PRIMARY KEY,
    account_id INT,
    amount DECIMAL(10,2),
    transaction_type ENUM('DEBIT','CREDIT'),
    transaction_date DATE,
    status ENUM('SUCCESS','FAILED'),
    FOREIGN KEY (account_id) REFERENCES accounts(account_id)
)
""")

print("âœ… transactions table created")
```

## 1.5 log table

```python
cursor.execute("""
CREATE TABLE logs (
    log_id INT PRIMARY KEY,
    user_id INT,
    log_level ENUM('INFO','WARN','ERROR'),
    message TEXT,
    created_at DATETIME,
    FOREIGN KEY (user_id) REFERENCES user(user_id)
)
""")

print("âœ… logs table created")
```
