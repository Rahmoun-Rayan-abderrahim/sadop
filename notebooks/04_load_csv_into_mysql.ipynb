## 4.4 Drop Existing Tables

We drop tables in reverse dependency order.


```python
users_df = pd.read_csv("../data/users.csv")
accounts_df = pd.read_csv("../data/Accounts.csv")
transactions_df = pd.read_csv("../data/transactions.csv")
logs_df = pd.read_csv("../data/logs.csv")
```

```python
user_query = """
INSERT INTO user (user_id, full_name, email, country, signup_date)
VALUES (%s, %s, %s, %s, %s)
"""

cursor.executemany(user_query, users_df.values.tolist())
conn.commit()

print(f"üë§ Inserted {len(users_df)} users")
```

```python
account_query = """
INSERT INTO accounts (account_id, user_id, account_type, balance, created_at)
VALUES (%s, %s, %s, %s, %s)
"""

cursor.executemany(account_query, accounts_df.values.tolist())
conn.commit()

print(f"üè¶ Inserted {len(accounts_df)} accounts")
```

```python
transaction_query = """
INSERT INTO transactions (transaction_id, account_id, amount, transaction_type, transaction_date,status)
VALUES (%s, %s, %s, %s, %s, %s)
"""

cursor.executemany(transaction_query, transactions_df.values.tolist())
conn.commit()

print(f"üí∏ Inserted {len(transactions_df)} transactions")
```

```python
log_query = """
INSERT INTO logs (log_id, user_id, log_level, message, created_at)
VALUES (%s, %s, %s, %s, %s)
"""

cursor.executemany(log_query, logs_df.values.tolist())
conn.commit()

print(f"üìú Inserted {len(logs_df)} logs")
```

```python
for table in ["user", "accounts", "transactions", "logs"]:
    cursor.execute(f"SELECT COUNT(*) FROM {table}")
    count = cursor.fetchone()[0]
    print(f"{table}: {count} rows")
```
