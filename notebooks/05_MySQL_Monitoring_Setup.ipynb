# 5. MySQL Monitoring Setup 


```python
import mysql.connector
from faker import Faker
import random
from datetime import datetime, timedelta
import pandas as pd
print("All set up")
```

## 5.1 Connect to MySQL


```python
conn = mysql.connector.connect(
    host="127.0.0.1",
    port=3307,
    user="sadop_user",
    password="1234",
    database="sadop_DBB"
)
cursor = conn.cursor()
print("Connected to SADOP database")
```

## 5.2 Check Current Slow Query Log Status


```python
cursor.execute("SHOW VARIABLES LIKE 'slow_query_log';")
print("slow_query_log:", cursor.fetchall())

cursor.execute("SHOW VARIABLES LIKE 'slow_query_log_file';")
print("slow_query_log_file:", cursor.fetchall())

cursor.execute("SHOW VARIABLES LIKE 'long_query_time';")
print("long_query_time:", cursor.fetchall())
```

## 5.3 Enable Slow Query Log


```python
cursor.execute("SET GLOBAL slow_query_log = 'ON';")
print("Slow query log enabled")
```

## 6.4 Set Slow Query Threshold


```python
# Set the slow query threshold to 0.7 seconds
cursor.execute("SET GLOBAL long_query_time = 0.7;")

# Verify immediately
cursor.execute("SHOW VARIABLES LIKE 'long_query_time';")
result = cursor.fetchall()
print("long_query_time:", result)
```

## 6.5 Verify Configuration


```python
cursor.execute("SHOW VARIABLES LIKE 'slow_query_log';")
print("slow_query_log:", cursor.fetchall())

cursor.execute("SELECT @@GLOBAL.long_query_time;")
print("Global long_query_time:", cursor.fetchone())

cursor.execute("SHOW VARIABLES LIKE 'slow_query_log_file';")
print("slow_query_log_file:", cursor.fetchall())
```
