# 9. Simulate Queries and Capture Metrics


## 9.1 Load Tables from CSV

```python
import pandas as pd

# Paths
users_csv = r"C:\Users\pc\data science\SADOP\data\Users.csv"
accounts_csv = r"C:\Users\pc\data science\SADOP\data\Accounts.csv"
transactions_csv = r"C:\Users\pc\data science\SADOP\data\Transactions.csv"
logs_csv = r"C:\Users\pc\data science\SADOP\data\Logs.csv"

# Load data
df_users = pd.read_csv(users_csv)
df_accounts = pd.read_csv(accounts_csv)
df_transactions = pd.read_csv(transactions_csv)
df_logs = pd.read_csv(logs_csv)

print("Tables loaded:")
print("Users:", len(df_users))
print("Accounts:", len(df_accounts))
print("Transactions:", len(df_transactions))
print("Logs:", len(df_logs))
```

## 9.2 Generate Synthetic Queries

```python
import time
import random

query_metrics = []

# Simulate 10000 queries
for _ in range(10000):
    start = time.time()
    
    # Random query example: sum of transactions for random account
    account_id = random.choice(df_accounts['account_id'].values)
    result = df_transactions[df_transactions['account_id'] == account_id]['amount'].sum()
    
    end = time.time()
    exec_time = end - start
    
    query_metrics.append({
        "query": f"SUM transactions for account_id={account_id}",
        "query_time": exec_time,
        "rows_examined": len(df_transactions[df_transactions['account_id'] == account_id])
    })

# Convert to DataFrame
df_metrics = pd.DataFrame(query_metrics)

df_metrics.shape
```

## 9.3 Save Metrics to CSV


```python
metrics_csv_path = r"C:\Users\pc\data science\SADOP\data\query_metrics.csv"
df_metrics.to_csv(metrics_csv_path, index=False)
print(f"Metrics saved to {metrics_csv_path}")
```
