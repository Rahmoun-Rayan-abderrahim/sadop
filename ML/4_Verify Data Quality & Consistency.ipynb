# 14 Verify Data Quality & Consistency

## 14.1 Load Dataset

```python
import pandas as pd

feature_dataset_path = r"C:\Users\pc\data science\SADOP\data\ml_features.csv"
df = pd.read_csv(feature_dataset_path)
df.head()
```

## 14.2 checking for missing values 

```python
# Check for nulls
missing = df.isnull().sum()
print("Missing values per column:\n", missing)
```

## 4.3 cheking for duplicated 

```python
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)
```

## 4.4 Duplicate Removal

```python
df_clean = df.drop_duplicates().reset_index(drop=True)
print("New shape:", df_clean.shape)
```

```python
df_clean.to_csv(r'C:\Users\pc\data science\SADOP\data\ml_features_clean.csv', index=False)
print(f"✅ ML features dataset verified and saved: ml_features_clean.csv")
```
